### DATASET: STUDENT PERFORMANCE DATASET

#### GOAL : SHOWING THE EFFECT OF STUDENTS PARENTS BACK GROUND, TEST PREPARATION. etc ON STUDENT PERFORMANCE
#### FIRST LOAD DATA

```python
#IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#LOAD DATA
df = pd.read_csv('datasets_74977_169835_StudentsPerformance (1).csv')
df.head()
```

```python
df.describe()
```

#### CREATE STUDENT PERFORMANCE COLUMN AND VIEW IT

```python
#create column called student performance with average of math, reading and writing score
#So that we can compare some columns with the newly creeated student performance column

performance_columns = df.loc[: , "math score":"writing score"]

df['student_performance']= performance_columns.mean(axis=1)

df['student_performance'].head(15)
df.head(5)
```

#### VALUE COUNT OF COLUMNS

```python
#value counts

print (df['parental level of education'].value_counts())

print (df['race/ethnicity'].value_counts())


print (df['lunch'].value_counts())


print (df['test preparation course'].value_counts())


print (df['gender'].value_counts())

```

# VISUALIZATIONS

# 1. SEABORN RELPLOTS

### 1(a) comparing Mathscore , student performace and testpreparation course

```python
#seaborn visualization

import seaborn as sns
#Mathscore vs student performace for testpreparation course

sns.relplot('math score','student_performance', data=df,hue='test preparation course')
```

### 1(b) comparing Mathscore, student performace, and parental level of education

```python

#seaborn visualization
#Mathscore vs student performace for parental level of education

sns.relplot('math score','student_performance', data=df,hue='parental level of education', style='parental level of education')
```

# 2. SEABORN LINE CHARTS

## 2(a) TEST PREPARATION COURSE

```python

#LINECHARTS

#Reading score
#seaborn visualization
#readingscore  vs student performace for testpreparation course
sns.relplot('reading score','student_performance', data=df,hue='test preparation course',kind='line',ci=None,col='test preparation course')
```

#### OBSERVATIONS: STUDENT PERFORMED BETTER WHEN THE TEST PREPARATION WAS COMPLETED FOR READING

```python
#MOST IMPORTANT

#LINECHARTS
#writing score
#seaborn visualization
#writing score vs student performace for testpreparation course
sns.relplot('writing score','student_performance', data=df,hue='test preparation course',kind='line',ci=None,col='test preparation course')
```

#### OBSERVATIONS: STUDENT PERFORMED BETTER WHEN THE TEST PREPARATION WAS COMPLETED FOR WRITING

```python
#LINECHARTS
#Maths score
#seaborn visualization
#Mathscore vs student performace for testpreparation course
sns.relplot('math score','student_performance', data=df,hue='test preparation course',kind='line',ci=None,col='test preparation course')
```

#### OBSERVATIONS: STUDENT PERFORMED BETTER WHEN THE TEST PREPARATION WAS COMPLETED FOR MATHS

## 2(b) PARENTAL EDUCATION

```python
#MOST IMPORTANT

#LINECHARTS
#parental education
#seaborn visualization
#Reading score vs student performace for parental level of education
sns.relplot('reading score','student_performance', data=df,hue='parental level of education',kind='line',ci=None,col='parental level of education',col_wrap=2)
```

# 3 SCATTERPLOT MOST EFFICIENT FOR THIS PROBLEM

### 3(a) TEST PREPARTION COURSE VS STUDENT PERFORMANCE

```python
#SCATTER PLOTS

#TEST PREPARTION COURSE VS STUDENT PERFORMANCE
sns.catplot("test preparation course","student_performance",data=df)
```

##### OBSERVATIONS MORE STUDENTS THAT COMPLETED THERE TEST PREPARATION FAILED MINIMUMLY AT 38, WHILE STUDENTS WiTH NO PREPARATION FAILED MINIMUMLY AT 4

## 3(b) PARENTAL LEVEL OF EDUCATION VS STUDENT PERFORMANCE

```python
#SCATTER PLOTS
#jitter=False

#parental level of education VS STUDENT PERFORMANCE
sns.catplot("parental level of education","student_performance",data=df, jitter=False)
```

## 3(c) RACE/ETHNICITY VS STUDENT PERFORMANCE

```python
#SCATTER PLOTS
#jitter=false
#race/ethnicity VS STUDENT PERFORMANCE
sns.catplot("race/ethnicity","student_performance",data=df,jitter=False)
```

## SCATTERPLOT WHEN SCORE IS LESS THAN 50

```python

#SCATTER PLOTS
#Query If student performance is less than 50
sns.catplot("test preparation course","student_performance",data=df.query("student_performance<50"),jitter=False)
```

## 4. BOX PLOTS

```python
#To see the most density of students with high performance based on test preparation course


sns.catplot(x="test preparation course",y ="student_performance",data=df, kind="violin")
```

##### OBSERVATIONS: STUDENTS THAT COMPLETED THERE TEST PREPARATION ARE MORE BTW 80 AND 60 PASSMARKS, WHILE STUDENTS WiTH NO PREPARATION ARE MORE BTW 80 AND 45 PASSMARKS

## 5. BARCHARTS

```python
#To see the highest total score on test preparation course and student performance


sns.catplot(x="test preparation course",y ="student_performance",data=df, kind="bar")
```
